```python
import pyqrcode
import png
```

```python
url = pyqrcode.create('https://twitter.com/home?lang=fr')
url.svg('mytwitterQrcode.svg', scale=6)
url.eps('mytwitterQrcode.eps', scale=2)
print(url.terminal(quiet_zone=1))
```

En utilisant le module pypng, on peut rendre le Qr code sous forme d'image PNG.

```python
phone_Qr = pyqrcode.create('+330000000', error='L', version=2, mode='binary')
phone_Qr.png('phone number.png', scale=6, module_color=[0, 0, 0, 128], background=[0xff, 0xff, 0xcc])
phone_Qr.show()
```
